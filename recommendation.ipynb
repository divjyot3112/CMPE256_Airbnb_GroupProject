```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('listings.csv')
df.head()
```

```python
df = df[['id', 'name', 'room_type', 'accommodates', 'price', 'review_scores_rating', 'number_of_reviews', 'property_type', 'beds', 'bed_type', 'bathrooms', 'bedrooms']]
df.head()
```

```python
df = df[df['number_of_reviews'] > 10]  
df = df.drop(['number_of_reviews'], axis=1)
df.head()
```

```python
df = df.fillna(0)
```

```python
df['room_type'] = pd.get_dummies(df['room_type'])
df['property_type'] = pd.get_dummies(df['property_type'])
df['bed_type'] = pd.get_dummies(df['bed_type'])
df.head()
```

```python
df['price'] = df['price'].replace('[\$,)]','', regex=True).replace('[(]','-', regex=True).astype(float)
df.head()
```

```python
df_features = df[['room_type', 'accommodates', 'price', 'review_scores_rating', 'property_type', 'beds', 'bed_type', 'bathrooms', 'bedrooms']]
df['name'] = df['name'].map(lambda name:re.sub('[^A-Za-z0-9]+', " ", name))
df_features.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
df_features = min_max_scaler.fit_transform(df_features)
```

```python
np.round(df_features,2)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
knn = NearestNeighbors(n_neighbors=7, algorithm='ball_tree').fit(df_features)
```

```python
distances, indices = knn.kneighbors(df_features)
```

```python
def print_similar_listings(query=None,id=None):
    for id in indices[id]:
            print(df.iloc[id])
            print('\n')
```
